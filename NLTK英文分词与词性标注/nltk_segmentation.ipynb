```python
import nltk, re, pprint
```

```python
# 从网上获取文本数据并对其进行分词
from urllib.request import urlopen

url = "http://www.gutenberg.org/files/2554/2554-0.txt"
raw = urlopen(url).read()
raw = str(raw)
tokens = nltk.word_tokenize(raw)
```

```python
print(type(tokens))
print(len(tokens))
```
