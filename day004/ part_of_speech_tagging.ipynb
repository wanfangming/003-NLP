```python
import jieba.posseg as psg
```

```python
# 读取txt文档改进版
filename = '围城.txt'
def get_txt_content(filename):
    with open(filename, 'r', encoding='utf-16', errors='ignore') as f:
        text = ''
        for l in f:
            l = l.strip()
            text += l
        return text
sent = get_txt_content(filename)
```

```python
seg_list = psg.cut(sent)
```

```python
for i in seg_list:
    print(i)
```
